```python


sales_orders = dfs["Sales Orders"]
customers    = dfs["Customers"]
regions      = dfs["Regions"]
products     = dfs["Products"]
```

```python
# Ensure dates & numerics
sales_orders["order_date"] = pd.to_datetime(sales_orders["order_date"], errors="coerce")

num_cols = ["order_quantity", "unit_price", "line_total", "total_unit_cost"]
for c in num_cols:
    sales_orders[c] = (
        sales_orders[c].astype(str).str.replace(",", "", regex=False).astype(float)
    )

# Profit & margin at row level
sales_orders["profit"] = sales_orders["line_total"] - sales_orders["total_unit_cost"]
sales_orders["margin_pct"] = (sales_orders["profit"] / sales_orders["line_total"]) * 100

# Lookups for readable names
sales_with_products = sales_orders.merge(
    products[["index", "product_name"]],
    left_on="product_description_index",
    right_on="index",
    how="left"
)

sales_with_regions = sales_orders.merge(
    regions[["id", "name"]],
    left_on="delivery_region_index",
    right_on="id",
    how="left"
).rename(columns={"name": "region_name"})

sales_with_customers = sales_orders.merge(
    customers,  # uses your standardized column names
    left_on="customer_name_index",
    right_on=customers.columns[0],  # first column should be the customer index
    how="left"
)
# Rename the customer name column for convenience (adjust if your header differs)
if "Customer Names" in sales_with_customers.columns:
    sales_with_customers = sales_with_customers.rename(columns={"Customer Names":"customer_name"})
elif "customer_names" in sales_with_customers.columns:
    sales_with_customers = sales_with_customers.rename(columns={"customer_names":"customer_name"})
```

```python
def margin_table(df, by, value_cols=("profit","line_total")):
    agg = df.groupby(by)[list(value_cols)].sum()
    agg["margin_pct"] = (agg["profit"] / agg["line_total"]) * 100
    return agg.sort_values("margin_pct", ascending=False)

channel_margin  = margin_table(sales_orders, "channel")
region_margin   = margin_table(sales_with_regions, "region_name")
product_margin  = margin_table(sales_with_products, "product_name")
warehouse_margin= margin_table(sales_orders, "warehouse_code")

# Side-by-side view (top 10 each)
display(channel_margin.head(10))
display(region_margin.head(10))
display(product_margin.head(10))
display(warehouse_margin.head(10))
```

```python
overall_margin = (sales_orders["profit"].sum()/sales_orders["line_total"].sum())*100

def high_rev_low_margin(df, by, top_n=15):
    agg = df.groupby(by)[["line_total","profit"]].sum()
    agg["margin_pct"] = (agg["profit"]/agg["line_total"])*100
    # flag low margin vs overall
    flagged = agg.sort_values("line_total", ascending=False).head(200)  # look at top revenue slice
    return flagged[flagged["margin_pct"] < overall_margin].sort_values(
        ["line_total","margin_pct"], ascending=[False, True]
    ).head(top_n)

print(f"Overall margin: {overall_margin:.2f}%")

print("\n⚠️ High revenue, low margin — PRODUCTS")
display(high_rev_low_margin(sales_with_products, "product_name"))

print("\n⚠️ High revenue, low margin — REGIONS")
display(high_rev_low_margin(sales_with_regions, "region_name"))

print("\n⚠️ High revenue, low margin — CHANNELS")
display(high_rev_low_margin(sales_orders, "channel"))

print("\n⚠️ High revenue, low margin — WAREHOUSES")
display(high_rev_low_margin(sales_orders, "warehouse_code"))
```

```python
cust = (sales_with_customers
        .groupby("customer_name")[["line_total","profit"]]
        .sum()
        .sort_values("line_total", ascending=False))

cust["revenue_share"] = cust["line_total"] / cust["line_total"].sum()
cust["cum_rev_share"] = cust["revenue_share"].cumsum()

print("Top 20 customers by revenue:")
display(cust.head(20))

# Pareto: % of customers needed to reach 80% revenue
pareto_cut = (cust["cum_rev_share"] <= 0.80).sum()
pct_of_customers = pareto_cut / len(cust) * 100
print(f"\nPareto result: Top {pareto_cut} customers ({pct_of_customers:.1f}% of customers) generate 80% of revenue.")

# Optional Pareto plot
plt.figure(figsize=(10,4))
cust["cum_rev_share"].reset_index(drop=True).plot()
plt.axhline(0.8, ls="--")
plt.title("Customer Pareto (Cumulative Revenue Share)")
plt.ylabel("Cumulative share")
plt.xlabel("Customers (sorted by revenue)")
plt.show()
```

```python
check_cols = ["unit_price","order_quantity","profit"]

def iqr_flags(s):
    q1, q3 = s.quantile([0.25,0.75])
    iqr = q3 - q1
    low  = q1 - 1.5*iqr
    high = q3 + 1.5*iqr
    return (s < low) | (s > high)

outlier_mask = pd.DataFrame({c: iqr_flags(sales_orders[c]) for c in check_cols})
outlier_rows = sales_orders[outlier_mask.any(axis=1)].copy()

print(f"Outlier rows found: {len(outlier_rows)}")
display(outlier_rows[["order_number","order_date","channel","warehouse_code","unit_price","order_quantity","profit"]].head(20))

# Optional boxplots
for c in check_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=sales_orders[c])
    plt.title(f"Outliers — {c}")
    plt.show()
```

```python
cr = (sales_with_regions
      .groupby(["channel","region_name"])["line_total"]
      .sum()
      .reset_index())

pivot_cr = cr.pivot(index="channel", columns="region_name", values="line_total").fillna(0)

plt.figure(figsize=(14,5))
sns.heatmap(pivot_cr, linewidths=.5)
plt.title("Revenue Heatmap: Channel × Region")
plt.xlabel("Region")
plt.ylabel("Channel")
plt.show()
```

```python
# 2017 actual sales by product
sales_2017_products = sales_with_products[sales_with_products["order_date"].dt.year == 2017]

actual_by_product = (
    sales_2017_products.groupby("product_name")["line_total"]
    .sum()
    .reset_index()
    .rename(columns={"line_total": "actual_2017"})
)

# Merge with budgets
budgets_2017 = dfs["2017 Budgets"].rename(columns={"product_name":"product_name", "2017_budgets":"budget_2017"})

comp_products = (
    actual_by_product.merge(budgets_2017, on="product_name", how="left")
    .assign(variance=lambda x: x["actual_2017"] - x["budget_2017"],
            variance_pct=lambda x: x["variance"] / x["budget_2017"] * 100)
)

print("2017 Actual vs Budget by Product:")
display(comp_products.sort_values("variance", ascending=False).head(20))

# Plot
comp_products.set_index("product_name")[["actual_2017","budget_2017"]].plot(
    kind="bar", figsize=(14,5))
plt.title("2017 Actual vs Budget by Product")
plt.ylabel("Revenue ($)")
plt.show()
```
